```python
#Step 1 import libraries
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
url='https://raw.githubusercontent.com/SIslam007/Python-for-Biologists/refs/heads/main/PDH%20KINETICS_Line%20plot.csv'
columns_to_read = ['Substrate(uM)', 'WT', 'S264D', 'S266D', 'S264A', 'S266A']
df=pd.read_csv(url, usecols=columns_to_read)
df
```

```python
# Step 5: Plot the growth curve
plt.figure(figsize=(6,4))
plt.plot(df.iloc[:,0], df.iloc[:,1],marker='o', color='DarkSlateBlue', label=df.columns[1])
plt.plot(df.iloc[:,0], df.iloc[:,2],marker='s', color='DarkSlateGray', label=df.columns[2])
plt.plot(df.iloc[:,0], df.iloc[:,3],marker='d', color='crimson', label=df.columns[3])
plt.plot(df.iloc[:,0], df.iloc[:,4],marker='v', color='teal', label=df.columns[4])
plt.plot(df.iloc[:,0], df.iloc[:,5],marker='^', color='coral', label=df.columns[5])

plt.xlabel("Time (hours)")
plt.ylabel("OD600")
plt.title("Bacterial Growth Curve")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
