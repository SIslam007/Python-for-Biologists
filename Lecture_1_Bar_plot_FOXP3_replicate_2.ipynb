```python
# === Step 1: Import libraries ===
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
# === Step 2: Data ===
control = [250000, 270000, 240000]
autoimmune = [140000, 120000, 125000]
```

```python
# === Step 3: Calculate mean and standard deviation ===
mean_values = [np.mean(control), np.mean(autoimmune)]
std_values = [np.std(control, ddof=1), np.std(autoimmune, ddof=1)]  # sample SD (ddof=1)
```

```python
# === Step 4: Student's t-test ===
t_stat, p_value = ttest_ind(control, autoimmune)
print("t =", t_stat)
print("p =", p_value)
```

```python
# === Step 5: Plot bar chart with error bars ===
labels = ['Control', 'Autoimmune disease']
colors = ['skyblue', 'lightcoral']

plt.bar(labels, mean_values, yerr=std_values, color=colors, capsize=8) #width=1 ,alpha=0.7
plt.ylabel('FOXP3 Expression')
plt.title('FOXP3 Expression in Control vs Autoimmune Disease')
# === Step 6: Integrate significance asterisk inside plot ===
# Decide significance symbol
if p_value < 0.001:
    sig = '***'
elif p_value < 0.01:
    sig = '**'
elif p_value < 0.05:
    sig = '*'
else:
    sig = 'ns'

# Position for asterisk (between the two bars, slightly above them)
max_height = max(mean_values) + max(std_values)
plt.text(0.5, max_height * 1.02, sig, ha='center', va='bottom', fontsize=18, fontweight='bold')

# Optional: add connecting line between bars
plt.plot([0, 0, 1, 1],
         [max_height * 1.01, max_height * 1.015, max_height * 1.015, max_height * 1.01],
         lw=1.2, color='black')

plt.tight_layout()
plt.show()
```
